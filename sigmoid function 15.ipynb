```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define points and their classes
x_points = np.array([-2, -1, 1, 2])
y_labels = np.array([0, 0, 1, 1])  # 0 = Class -1, 1 = Class +1

# Define theta and theta0 combinations
parameters = [
    (1, 0),
    (-1, 0),
    (0.5, 0),
    (5, 0)
]

# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

for ax, (theta, theta0) in zip(axs.ravel(), parameters):
    # Generate x range and corresponding sigmoid values
    x_range = np.linspace(-3, 3, 300)
    z = theta * x_range + theta0
    y = sigmoid(z)

    # Plot the sigmoid curve
    ax.plot(x_range, y, label=f"Theta={theta}, Theta0={theta0}")

    # Plot decision threshold
    ax.axhline(y=0.5, color='gray', linestyle='--')

    # Plot the data points
    for x, label in zip(x_points, y_labels):
        color = 'red' if label == 0 else 'blue'
        ax.scatter(x, label, c=color, edgecolors='k')

    # Add labels and legend
    ax.set_title(f"Theta={theta}, Theta0={theta0}")
    ax.set_xlabel("x")
    ax.set_ylabel("Sigmoid Output")
    ax.grid(True)
    ax.legend(loc='best')

# Add custom legend for classes (manually)
handles = [
    plt.Line2D([0], [0], marker='o', color='w', label='Class -1', markerfacecolor='red', markeredgecolor='k'),
    plt.Line2D([0], [0], marker='o', color='w', label='Class +1', markerfacecolor='blue', markeredgecolor='k')
]
fig.legend(handles=handles, loc='upper center', ncol=2)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
